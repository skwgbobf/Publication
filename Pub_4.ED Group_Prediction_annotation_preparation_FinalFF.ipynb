### III. Annotation for NN, YN and YY

```python
from elastic_util import *
from elasticsearch import Elasticsearch
from elasticsearch_dsl import Q
from datetime import datetime

client = Elasticsearch('http://elastic:Admin2019@10.200.112.204:9200', 
                       request_timeout=600000, retry_on_timeout=True)

if not client.ping():
    raise ValueError("Connection failed")
```

```python
index = 'mimic_medcat'
doc_type = 'doc'
```

```python
# meta_anns.Negated.name, meta_anns.Negated.value, meta_anns.Skip.name, meta_anns.Skip.value
```

```python
def all_df_chunk(row_id_list, all_df=None, size=1000):
    must = [Q("match", meta_anns__Negated__value='No'), Q("match", meta_anns__Skip__value='No')]
    dfs = []

    data = row_id_list
    for chunk in [data[x:x+size] for x in range(0, len(data), size)]:
        should = []
        print('------------------')
    #     print(chunk)
        for row_id in chunk:
          should.append(Q("match", row_id=row_id))
        q = Q('bool',
            must=must,
            should=should,
            minimum_should_match=1
        )

        return_fields = ['pretty_name','cui', 'tui', 'type', 'source_value', 'row_id']

        s = Search(using=client, index=index, doc_type=doc_type).source(return_fields).filter(q)
        df = pd.DataFrame((d.to_dict() for d in s.scan()))
        print(datetime.now(), df.shape)
        dfs.append(df)
    all_df = pd.concat(dfs)
    return all_df
```
