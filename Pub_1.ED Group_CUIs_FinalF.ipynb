```python
# {id:'""" + cui + """' }
#  RETURN p.cui, p.name, """
def cui2name(cui):
    query = """
        MATCH (p:Concepts{id:'""" + cui + """' })
        RETURN p.name"""
    return (graph.run(query).to_series()).tolist()
```

```python
def cui2tui(cui):
    query = """
        MATCH (a:Types) <-[r:cui_2_tui]-(d:Concepts)
        WHERE d.id IN ['""" + cui + """']
        RETURN a.name"""
    return (graph.run(query).to_series()).tolist()
```

```python
def cui2id(cui):
    query = """
        MATCH (a:Types) <-[r:cui_2_tui]-(d:Concepts)
        WHERE d.id IN ['""" + cui + """']
        RETURN a.id"""
    return (graph.run(query).to_series()).tolist()
```

```python
def cui2scv(cui):
    query = """
        MATCH (sc: Terms)- [r:aui_2_cui]->(con:Concepts)
        WHERE con.id IN ['""" + cui + """']
        RETURN sc.vocabulary"""
    return (graph.run(query).to_series()).tolist()
```

```python
def cui2scot(cui):
    query = """
        MATCH (sc: Terms)- [r:aui_2_cui]->(con:Concepts)
        WHERE con.id IN ['""" + cui + """']
        RETURN sc.other_code"""
    return (graph.run(query).to_series()).tolist()
```

```python
def cui2scid(cui):
    query = """
        MATCH (sc: Terms)- [r:aui_2_cui]->(con:Concepts)
        WHERE con.id IN ['""" + cui + """']
        RETURN sc.id"""
    return (graph.run(query).to_series()).tolist()
```

```python
def cui2scname(cui):
    query = """
        MATCH (sc: Terms)- [r:aui_2_cui]->(con:Concepts)
        WHERE con.id IN ['""" + cui + """']
        RETURN sc.name"""
    return (graph.run(query).to_series()).tolist()
```

```python
#1. create cui data frame
column_names = ["cui", "row_id"]
cuis = pd.DataFrame(columns = column_names)
```

```python
# Annotation : extrcting all row_ids which have any terms mapped into cui_list
row_ids = set()
cui_ids =set()

for cui in cui_list:
    q = Q("match", cui=cui) & Q("match", meta_anns__Negated__value='No') & Q("match", meta_anns__Skip__value='No') 
    s = Search(using=client, index=index, doc_type=doc_type).filter(q)
    df = pd.DataFrame((d.to_dict() for d in s.scan()))
    if len(df) > 0:
        rids = df['row_id'].tolist()
        cus = df['cui'].tolist()
        i=pd.DataFrame(data=zip(df['cui'],df['row_id']), columns=['cui','row_id'])
        cuis=cuis.append(i, ignore_index=True)  
        row_ids.update(rids)
        cui_ids.update(cus)
```

```python
def all_df_chunk(row_id_list, all_df=None, size=1000):
    must = [Q("match", meta_anns__Negated__value='No'), Q("match", meta_anns__Skip__value='No')]
    dfs = []

    data = row_id_list
    for chunk in [data[x:x+size] for x in range(0, len(data), size)]:
        should = []
        print('------------------')
    #     print(chunk)
        for row_id in chunk:
          should.append(Q("match", row_id=row_id))
        q = Q('bool',
            must=must,
            should=should,
            minimum_should_match=1
        )

        return_fields = ['pretty_name','cui', 'tui', 'type', 'source_value', 'row_id']

        s = Search(using=client, index=index, doc_type=doc_type).source(return_fields).filter(q)
        df = pd.DataFrame((d.to_dict() for d in s.scan()))
        print(datetime.now(), df.shape)
        dfs.append(df)
    all_df = pd.concat(dfs)
    return all_df
```
